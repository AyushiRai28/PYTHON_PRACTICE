```python
# gcd using euclidean algorithm
# divident = divisor * qoutient + remainder 
def gcd(a,b):
    divident=a
    divisor = b
    if b>a:
        divident = b
        divisor = a
    if b==a:
        return a
    while(divisor>0):
        remainder = divident%divisor
        divident = divisor 
        divisor = remainder
    return divident
gcd(24,36)    
```

```python
# gcd using recursion
def gcd(a,b):
    if b==0:
        return a
    else :
        return gcd(b, a%b)
gcd(24,36)
        
```
