```python
#write a code to add all priime numbers, from a given input of 5 numbers 
i =  1 
sum = 0
while i<=5:
    n = int(input("enter number :"))
    for j in range(2 , int(n**0.5)):
        if n % j == 0:
            sum = sum+0
        else:    
            sum = sum +n        
    i = i +1        
print(sum)        
```

```python
#write a code to add all priime numbers, from a given input of 5 numbers 
import array as ar
i =  1 
a = ar.array('i', [0,0,0,0,0])
while i<=5:
    a[i] = int(input("enter number :"))
    for j in range(2 , int(a[i]**0.5)):
        if a[i] % j == 0:
            a[i] == 0   
    i = i +1  
sum = sum(a[i])   
print(sum)        
```

```python
#write a code to sort 10 numbers, withpour any in built function
import array as ar
def sort():
    a=ar.array('i', [0]*10)
    for i in range(0,10):
        a[i] = int(input("enter number:"))
        for i in range(0,10):
            for j in range(0,10):
                if a[j]>a[j+1]:
                    t = a[j+1]
                    a[j+1] = a[j]
                    a[j] = t
    for i in range(0,10):
        print(a[i])
    
    
```
