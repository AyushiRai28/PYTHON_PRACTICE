```python
#code to find forward and backward diagonal of martix
import numpy as np

a = np.zeros((3,3))
for i in range(0,3):
    for j in range(0,3):
        a[i][j] = int(input("Enter number:"))

for i in range(0,3):
    print(a[i][i])
    print(a[i][2-i])
```

```python
# code for matrix multiplication
import numpy as np
a = np.zeros((3,3))
b = np.zeros((3,3))
c = np.zeros((3,3))
for i in range(0,3):
    for j in range(0,3):
        for k in range(0,3):
            c[i][j] =  c[i][j] + a[i][kj]*b[k][j]

for i in range(0,3):
    print(' ')
    for j in range(0,3):
        print(c[i][j] , " ", end=" ")
```
