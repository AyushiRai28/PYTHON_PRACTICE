```python
#decimal to binary
def binary(n):
    sum=0
    i=0
    while n>0:
        r = n%2
        sum = sum + (r*10**i)
        i = i+1
        n = int(n/2)
    return sum        
binary(15)        
```

```python
# binary to decimal
def decimal(b):
    sum = 0
    count = 0
    while b>0:
       r=b%10
       sum=sum+ (r*2**count)
       count = count+1
       b= int(b/10)
    return sum
decimal(1111)
```
